```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
```

```python
# загрузка данных
dataset = pd.read_csv('diabetes.csv')
print(dataset.info())
```

```python
rows, cols = dataset.shape

print(f"В датасете {cols} колонок и {rows} строк")
```

```python
dataset.columns
```

```python
dataset.columns = [str.lower(col) for col in dataset.columns]
```

```python
dataset.columns
```

```python
has_diabet = dataset['outcome'] == 1
has_no_diabet = dataset['outcome'] == 0
# print(has_diabet) #Serias boolean + index 

diabetic_dataset = dataset[has_diabet]
# print(diabetic_dataset)

no_diabetic_dataset = dataset[has_no_diabet]
```

```python
rows01, cols01 = diabetic_dataset.shape

print(f"В датасете Диабет+ {cols01} колонок и {rows01} строк")
```

```python
rows02, cols02 = no_diabetic_dataset.shape

print(f"В датасете Диабет- {cols02} колонок и {rows02} строк")
```

```python
print(type(diabetic_dataset))
print(diabetic_dataset.info())
```

```python
plt.hist(x=diabetic_dataset['age'], color='green', edgecolor='white', bins=20)
plt.title("Распределение больных диабетом по возрасту")
plt.xlabel("Возраст (лет)")
plt.ylabel("Количество больных диабетом")
```

# EDA исследовательский анализ данных

```python
data_to_plot_age = [diabetic_dataset['age'], no_diabetic_dataset['age']]
# # print(data_to_plot)
# min = 0
# max = 80
# cnt = 10
# x_points = np.linspace(min, max, cnt)

# color = ['gray', 'blue']

# show_plot(data_to_plot, x_points, color)


x_points, step = np.linspace(20, 80, 7, retstep=True)
print (step)
plt.hist(data_to_plot_age, x_points, color = ['purple', 'lightblue'], label=['больные диабетом', 'здоровые'])
plt.legend(loc='upper right')
plt.xlabel("Возраст")
plt.ylabel("Количество пациентов")
plt.title("Статистика по возрасту для базы Диабет")
plt.grid(color="gainsboro") # Сетка
plt.show()
```

```python
dataset.describe()
```

```python
has_bmi = dataset['bmi'] > 0.0
group_features_test = dataset[has_bmi]
group_features = group_features_test.groupby(['bmi', 'age'], as_index=False).agg({'glucose': 'mean'})

group_features.head(15)
```

```python
px.bar(group_features, x='age', y='glucose', color='bmi')
```

```python
pd.plotting.scatter_matrix(group_features, figsize=(10,10), edgecolor='white', marker='o', hist_kwds = {'bins': 8}, s=100, alpha=1.0)
```

```python
sns.pairplot(group_features)
```

```python
plt.scatter(group_features['age'], group_features['bmi'])
plt.xlabel('Возраст')
plt.ylabel('Индекс массы тела')
plt.show()
```
